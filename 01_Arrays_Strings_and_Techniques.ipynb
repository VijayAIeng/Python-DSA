### 1. ARRAYS & STRINGS

* Array (List in Pytho) a collection of items stored in contiguous memory location

* String : A sequence of char like "vijay" , strings are immutable (cannot change individual char directly) while List are mutable .

```python
#Traversal -> Visiting Each elements (anytime you need to look or modify every element (sum,count,find,etc))

a = [10,20,30,40,50,60,70,90]

for i in a:
  print(i)
```

```python
#searching -> fine the specific element (Looking for presence or position of element)

a = [10,20,30,40,50,60,70,90]

find = 60

for i in range(len(a)):
    if a[i] == find :
       print(f"Found {find} at index {i}")
       #break
```

```python
#inserting -> adding new element

a = [10,20,30,40,50,60,70,90]

a.append(100) #add at end
a.insert(1,101) #insert 101 at index 1 (insert(index,element))
print(a)
```

```python
#Deletion -> remove element

a= [10, 101, 20, 30, 40, 50, 60, 70, 90, 100]

a.remove(101) #remove by value
del a[4]  #remove by index 30

print(a)
```

### 2. Two-pointer techniques

* use the 2 indics(pointers) to move through the list(array) different direction to solve problem effeciently

```python
""" Given a sorted array of integers a and a target sum find,
find two numbers in the array whose sum is exactly equal to find.

Return or print those two numbers."""


a = [10, 20, 30, 40, 60, 70, 90, 100]
find = 70

left,right = 0,len(a)-1

while left < right:
    current_sum = a[left] +a[right]
    if current_sum == find:
        print(a[left],a[right])
        break

    elif current_sum < find:
        left += 1
    else:
        right -= 1
```

### 3. Sliding Window technique

Looking at a few items (a group) at a time, then moving forward step by step.

Sliding window = find the best continuous group without redoing all the work

* Find the maximum number of fruits you can get from any 3 consecutive boxes.



```python
boxes = [4, 2, 10, 3, 12, 5, 8]
k = 3  # number of boxes you can carry

window_sum = sum(boxes[:k])  # initial window sum
max_sum = window_sum #storing max values

for i in range(k,len(boxes)):
    print(boxes[i])
    window_sum += boxes[i] - boxes[i-k]
    max_sum = max(max_sum, window_sum)

print("Maximum fruits you can carry =", max_sum)
```
