##DATA TYPES WITH MEMORY ALLOCATION

###Immutable datatypes (memory efficient)

```python
#str

name = "python"
print(f"string memory:{name.__sizeof__()} bytes")
```

```python
#int

num = 4367
print(f"number memory:{num.__sizeof__()} bytes")
```

```python
#tuple

coordinates = (19,21,34)
print(f"coordinates memory:{coordinates.__sizeof__()}  bytes")
```

### Mutable Types (Dynamic Memory)

```python
# List - Dynamic array with overallocation

LIST = [5,4,3,2,1]
print(f"LIST memory:{LIST.__sizeof__()}  bytes")
```

```python
## Dictionary - Hash table with load factor

user_data = {"name": "vijay", "age": 27}
print(f"Dict memory: {user_data.__sizeof__()} bytes")
```

```python
###__slots__ to reduce dict memory with class

class Memorry_Efficient:
  __slots__=['name','age']
  def __init__(self, name , age ):
      self.name =  name
      self.age = age


class RegularClass:
    def __init__(self, name, age):
        self.name = name
        self.age = age


regular_class = RegularClass("ajith kumar",27)
Memorry_class = Memorry_Efficient("vijay ",27)

print(f"Regular: {regular_class.__sizeof__()} bytes")
print(f"Slotted: {Memorry_class.__sizeof__()} bytes")

"""
Regular: 16 bytes → __sizeof__() shows only the object header, hiding the __dict__ which lives separately (adding ~48+ bytes)

Slotted: 32 bytes → __sizeof__() shows the complete object because slots store attributes directly inside, with no hidden dictionary

Regular: 16 bytes
Slotted: 32 bytes

"""
```

```python
"""
COMPARISON CHART (for n = 1000 items)
====================================
O(1):       1 operation     (Instant)
O(log n):   ~10 operations  (Very fast)
O(n):       1000 operations (Average)
O(n log n): 10,000 ops      (Slow)
O(n²):      1,000,000 ops   (Very slow)
"""

def visualize_complexities():
    print("=== HOW FAST DO ALGORITHMS GROW? ===\n")

    n = 10  # Let's see with 10 items

    print(f"For {n} items:")
    print(f"O(1):      {1:>10} operations")
    print(f"O(log n):  {len(bin(n))-2:>10} operations")
    print(f"O(n):      {n:>10} operations")
    print(f"O(n²):     {n*n:>10} operations")
    print(f"O(2ⁿ):     {2**n:>10} operations")

    print("\nFor 1000 items (Imagine the difference!):")
    n = 1000
    print(f"O(1):      {1:>10} operations")
    print(f"O(log n):  {len(bin(n))-2:>10} operations")
    print(f"O(n):      {n:>10} operations")
    print(f"O(n²):     {n*n:>10} operations")
    print(f"O(2ⁿ):     {2**n:>10} operations (IMPOSSIBLE!)")

visualize_complexities()
```
